```python
#FETCH SOME DATA
url = 'https://pour-app.s3.amazonaws.com/'
response = requests.get(url)
z = response.content
x = xmltodict.parse(z)
N_min = 20000
for k in x['ListBucketResult']['Contents']:
    if int(k['Size']) > N_min:
        fname = '%s%s' % (url, k['Key']) 
        path = '%s/%s' % (PATH_TO_DATA,k['Key'])
        urllib.urlretrieve(fname,path)
        
        
        
```

```python
chart = nvd3.lineWithFocusChart(name="focusChart",height=450)

for filename in os.listdir(PATH_TO_DATA):
    if '.json' in filename:

        dfile = json.load(open('%s/%s' % (PATH_TO_DATA, filename)))
        ddict = [{'time':k[0], 'mass':k[1]} for k in dfile]
        df = pd.DataFrame(ddict)
        xdata = df.time
        ydata = df.mass
        chart.add_serie(x=xdata,y=ydata,name=filename)

```

```python
chart.buildhtml()
output_file = open('coffee_pl.html', 'w')
output_file.write(chart.htmlcontent)
output_file.close()

```
